```python
# Importando as bibliotecas necessárias
import pandas as pd
import numpy as np
import os
```

```python
# Leitura de dados

data_dir = r"C:\Users\x432601\Downloads\SIMULACAO_MONTE_CARLO\data_testes\data"
file01 = r"calculo-emissao-poluentes-diario_2026-01-01_silver.csv"

df_consume_table = pd.read_csv(os.path.join(data_dir, file01), sep = ",")
```

```python
df_consume_table.head()
```

```python
df_consume_table.describe()
```

```python
# Renomeando e trocando dimensão das colunas de emissões para toneladas (t)
df_consume_table = df_consume_table.rename(columns = {'emissao_co2': 'emissao_co2(t)', 'emissao_mp' : 'emissao_mp(t)', 'emissao_nox':'emissao_nox(t)' })
df_consume_table['emissao_co2(t)'] = df_consume_table['emissao_co2(t)'] / 1000 # Convertendo de kg para toneladas
```

```python
# Salvando o arquivo corrigido
output_file = r"calculo-emissao-poluentes-diario_2026-01-01_silver_corrigido.csv"
df_consume_table.to_csv(os.path.join(data_dir, output_file), index = False)
```
